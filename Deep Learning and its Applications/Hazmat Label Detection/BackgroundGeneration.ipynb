This code below generates background images from the Flickr8k Dataset.
Each file saved represents 64 different background images. Each .npy file represent a batch of shape (64, 150, 150, 3).

```python
from skimage.io import imread, imshow, imsave
import numpy as np
from os import listdir
from math import floor
```

```python
bg_names = listdir("Flicker8k")
num_bg=len(bg_names)
bgs=[]
stride = 16
c = 0
d = 0
f = 0
for path in bg_names:
    pic = imread("Flicker8k/{}".format(path))
    s = pic.shape
    I = floor((s[0] - 150) / stride)
    J = floor((s[1] - 150) / stride)
    print(c)
    c+=1
    for i in range(I):
        for j in range(J):
            _ = pic[i*stride:i*stride+150,j*stride:j*stride+150]
            bgs.append(_)
            if d % 64 == 0:
                np.save("bgs/{}".format(f), np.array(bgs))
                bgs = []
                f += 1
            d += 1
```
